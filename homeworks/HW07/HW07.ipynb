```python
def load_dataset(path, id_col="sample_id"):
    df = pd.read_csv(path)

    print("head")
    display(df.head())

    print("\n\ninfo")
    display(df.info())

    print("\n\ndescribe")
    display(df.describe(include="all"))

    print("\n\nКоличество пропусков")
    display(df.isna().sum())

    print("\n\nДоля пропусков")
    display(df.isna().mean())

    # Типы признаков
    numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
    categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()

    # Удаляем sample_id, если он случайно попал в списки
    if id_col in numeric_cols:
        numeric_cols.remove(id_col)
    if id_col in categorical_cols:
        categorical_cols.remove(id_col)

    print("\nЧисловые признаки:", numeric_cols)
    print("Категориальные признаки:", categorical_cols)

    return df, numeric_cols, categorical_cols
```

```python
def split_features(df, id_col="sample_id"):
    """
    Разделяет датафрейм на:
    - X: все признаки, кроме id_col
    - sample_ids: колонка для сохранения результатов
    """

    X = df.drop(columns=[id_col])
    sample_ids = df[id_col]

    print("Признаки X")
    display(X.head())

    print("\nsample_id")
    display(sample_ids.head())

    print("\nРазмерности:")
    print("X:", X.shape, " sample_id:", sample_ids.shape)

    return X, sample_ids
```

#### `Препроцессинг`

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

Дальше будет написана функция для каждого датасета, где нужно контролировать флажочек есть ли категориальные признаки (можно было конечно это оптимезировать, но мне пока лень)

```python
def build_preprocessor(df, numeric_cols, categorical_cols=None, encode_categorical=False):
    """
    Препроцессинг для датасета:
    - проверяет пропуски
    - числовые: SimpleImputer + StandardScaler
    - категориальные: SimpleImputer + OneHotEncoder (если encode_categorical=True) (в данном случае для dataset-04)
    """
    # Проверка пропусков
    print("Проверка пропусков:")
    print(df[numeric_cols + (categorical_cols or [])].isna().sum())

    transformers = []

    # Числовые признаки
    print("\nЧисловые признаки:", numeric_cols)
    numeric_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])
    transformers.append(("num", numeric_transformer, numeric_cols))

    # Категориальные признаки
    if categorical_cols:
        print("Категориальные признаки:", categorical_cols)

        if encode_categorical:
            print("Категориальные признаки будут закодированы OneHotEncoder.")
            categorical_transformer = Pipeline(steps=[
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("encoder", OneHotEncoder(handle_unknown="ignore"))
            ])
            transformers.append(("cat", categorical_transformer, categorical_cols))
        else:
            print("Категориальные признаки НЕ кодируются и будут исключены из модели.")

    preprocessor = ColumnTransformer(transformers=transformers)
    return preprocessor
```

```python
preprocessor_1 = build_preprocessor(
    df=df1,
    numeric_cols=num_cols1,
    categorical_cols=cat_cols1,
    encode_categorical=False # нет категориальных признаков
)

X1_processed = preprocessor_1.fit_transform(X1)
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN 
from sklearn.cluster import AgglomerativeClustering
```

KMeans

```python
def run_kmeans(X, k_range, dataset_name):
    results = []

    for k in k_range:
        model = KMeans(
            n_clusters=k,
            random_state=42,
            n_init=10
        )
        labels = model.fit_predict(X)

        # Метрики
        sil = silhouette_score(X, labels)
        db = davies_bouldin_score(X, labels)
        ch = calinski_harabasz_score(X, labels)

        results.append({
            "k": k,
            "silhouette": sil,
            "davies_bouldin": db,
            "calinski_harabasz": ch,
            "labels": labels
        })

    # График silhouette vs k
    plt.figure(figsize=(6, 4))
    plt.plot([r["k"] for r in results], [r["silhouette"] for r in results], marker="o")
    plt.title(f"KMeans: зависимость silhouette от k ({dataset_name})")
    plt.xlabel("k")
    plt.ylabel("silhouette")
    plt.grid(True)
    plt.show()

    # Выбор лучшего k
    best = max(results, key=lambda r: r["silhouette"])
    print(f"Лучший k для {dataset_name}: {best['k']} (silhouette={best['silhouette']:.4f})")

    return results, best
```

DBSCAN

```python
def run_dbscan(X, eps_list, min_samples_list, dataset_name):
    results = []

    for eps in eps_list:
        for ms in min_samples_list:
            model = DBSCAN(eps=eps, min_samples=ms)
            labels = model.fit_predict(X)

            # Шум
            noise_ratio = np.mean(labels == -1)

            # Если все точки шум — метрики не считаем
            if len(set(labels)) <= 1 or len(set(labels)) == 2 and -1 in set(labels):
                sil = None
                db = None
                ch = None
            else:
                mask = labels != -1
                sil = silhouette_score(X[mask], labels[mask])
                db = davies_bouldin_score(X[mask], labels[mask])
                ch = calinski_harabasz_score(X[mask], labels[mask])

            results.append({
                "eps": eps,
                "min_samples": ms,
                "silhouette": sil,
                "davies_bouldin": db,
                "calinski_harabasz": ch,
                "noise_ratio": noise_ratio,
                "labels": labels
            })

    # Пример графика silhouette vs eps (для фиксированного min_samples)
    fixed_ms = min_samples_list[0]
    subset = [r for r in results if r["min_samples"] == fixed_ms and r["silhouette"] is not None]

    if subset:
        plt.figure(figsize=(6, 4))
        plt.plot([r["eps"] for r in subset], [r["silhouette"] for r in subset], marker="o")
        plt.title(f"DBSCAN: зависимость silhouette от eps (min_samples={fixed_ms}, {dataset_name})")
        plt.xlabel("eps")
        plt.ylabel("silhouette")
        plt.grid(True)
        plt.show()

    # Выбор лучшего
    valid = [r for r in results if r["silhouette"] is not None]
    if valid:
        best = max(valid, key=lambda r: r["silhouette"])
        print(f"Лучший DBSCAN для {dataset_name}: eps={best['eps']}, min_samples={best['min_samples']}, silhouette={best['silhouette']:.4f}")
    else:
        best = None
        print(f"DBSCAN не нашёл валидных кластеров для {dataset_name}")

    return results, best
```

#### `Метрики качества`

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
```

```python
def schet_metric(X, labels, model_name="model"):
    labels = np.array(labels)
    unique_labels = set(labels)

    # Доля шума
    noise_ratio = np.mean(labels == -1) if -1 in unique_labels else 0.0
    print(f"\nМетрики для {model_name}")
    print(f"Доля шума: {noise_ratio:.4f}")

    # Если есть шум — считаем только по non-noise
    if -1 in unique_labels:
        mask = labels != -1
        X_valid = X[mask]
        labels_valid = labels[mask]
        print("Метрики считаются по non-noise точкам (DBSCAN).")
    else:
        X_valid = X
        labels_valid = labels

    # Метрики
    sil = silhouette_score(X_valid, labels_valid)
    db = davies_bouldin_score(X_valid, labels_valid)
    ch = calinski_harabasz_score(X_valid, labels_valid)

    print(f"Silhouette: {sil:.4f}")
    print(f"Davies-Bouldin: {db:.4f}")
    print(f"Calinski-Harabasz: {ch:.4f}")

    return {
        "silhouette": sil,
        "davies_bouldin": db,
        "calinski_harabasz": ch,
        "noise_ratio": noise_ratio
    }
```

```python
# Для KMeans
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
labels_km = kmeans.fit_predict(X1_processed)
metrics_km = schet_metric(X1_processed, labels_km, model_name="KMeans")


# Для DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels_db = dbscan.fit_predict(X1_processed)
metrics_db = schet_metric(X1_processed, labels_db, model_name="DBSCAN")
```

```python
def build_filename(dataset=None, method=None, suffix=None):
    parts = []
    if dataset is not None:
        parts.append(dataset)
    if method is not None:
        parts.append(method)
    if suffix is not None:
        parts.append(suffix)
    return "_".join(parts) + ".png"
```

Дальше же написала функции для создания графиков

```python
def plot_pca(X, labels, title, dataset=None, method=None, suffix="pca"):
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X)

    plt.figure(figsize=(6, 5))
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap="tab10", s=12)
    plt.title(title)
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.colorbar(scatter, label="Кластер")
    plt.grid(True)

    if dataset is not None:
        filename = build_filename(dataset, method, suffix)
        plt.savefig(f"artifacts/figures/{filename}", dpi=200)

    plt.show()
```

```python
def plot_vs_k(results, dataset_name, dataset=None, suffix="silhouette_vs_k"):
    ks = [r["k"] for r in results]
    sils = [r["silhouette"] for r in results]

    plt.figure(figsize=(6, 4))
    plt.plot(ks, sils, marker="o")
    plt.title(f"Silhouette vs k ({dataset_name})")
    plt.xlabel("k")
    plt.ylabel("silhouette")
    plt.grid(True)

    if dataset is not None:
        filename = build_filename(dataset, None, suffix)
        plt.savefig(f"artifacts/figures/{filename}", dpi=200)

    plt.show()
```

```python
def plot_vs_eps(results, min_samples, dataset_name, dataset=None, suffix="silhouette_vs_eps"):
    subset = [r for r in results if r["min_samples"] == min_samples and r["silhouette"] is not None]
    eps_values = [r["eps"] for r in subset]
    sils = [r["silhouette"] for r in subset]

    plt.figure(figsize=(6, 4))
    plt.plot(eps_values, sils, marker="o")
    plt.title(f"Silhouette vs eps (min_samples={min_samples}, {dataset_name})")
    plt.xlabel("eps")
    plt.ylabel("silhouette")
    plt.grid(True)

    if dataset is not None:
        filename = build_filename(dataset, None, suffix)
        plt.savefig(f"artifacts/figures/{filename}", dpi=200)

    plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
# 5 разных random_state
states = [0, 1, 2, 3, 4]

# сохраняем разбиения
labels_list = []

for rs in states:
    km = KMeans(n_clusters=best_k_3, random_state=rs, n_init=10)
    labels = km.fit_predict(X3_processed)
    labels_list.append(labels)

# считаем ARI между всеми парами запусков
ari_matrix = np.zeros((5, 5))

for i in range(5):
    for j in range(5):
        ari_matrix[i, j] = adjusted_rand_score(labels_list[i], labels_list[j])

ari_matrix
```
